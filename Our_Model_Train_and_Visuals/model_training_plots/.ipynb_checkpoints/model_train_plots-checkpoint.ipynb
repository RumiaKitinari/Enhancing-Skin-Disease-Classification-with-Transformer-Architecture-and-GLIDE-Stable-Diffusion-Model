```python
import matplotlib.pyplot as plt
import os

def plot_all_histories(history_dict, save_dir="plots"):
    os.makedirs(save_dir, exist_ok=True)

    for model_name, datasets in history_dict.items():
        for dataset_name, history in datasets.items():

            epochs = history["epochs"]
            training_loss = history["training_loss"]
            validation_loss = history["validation_loss"]
            accuracy = history["accuracy"]
            final_accuracy = history.get("final_accuracy", None)

            plt.figure(figsize=(12, 4))

            # --------------------
            # Loss subplot
            # --------------------
            plt.subplot(1, 2, 1)
            plt.plot(
                epochs, training_loss,
                marker='o', linewidth=2,
                color='#1f77b4', label='Training Loss'
            )
            plt.plot(
                epochs, validation_loss,
                marker='s', linewidth=2,
                color='#d62728', label='Validation Loss'
            )

            plt.xlabel("Epoch", fontsize=11)
            plt.ylabel("Loss", fontsize=11)
            plt.title(f"{model_name.upper()} | {dataset_name.upper()} Loss", fontsize=12, weight='bold')
            plt.grid(True, linestyle='--', alpha=0.5)
            plt.legend(frameon=True)

            # --------------------
            # Accuracy subplot
            # --------------------
            plt.subplot(1, 2, 2)
            plt.plot(
                epochs, accuracy,
                marker='o', linewidth=2,
                color='#2ca02c', label='Validation Accuracy'
            )

            # --- Add final accuracy point ---
            if final_accuracy is not None and len(epochs) > 0:
                final_epoch = epochs[-1]

                plt.scatter(
                    final_epoch, final_accuracy,
                    color='red', s=80, zorder=5,
                    label='Test Accuracy'
                )

                # optional annotation
                plt.annotate(
                    f"{final_accuracy:.3f}",
                    (final_epoch, final_accuracy),
                    textcoords="offset points",
                    xytext=(-15, -15),
                    fontsize=9,
                    color='red'
                )

            plt.xlabel("Epoch", fontsize=11)
            plt.ylabel("Accuracy", fontsize=11)
            plt.title(f"{model_name.upper()} | {dataset_name.upper()} Accuracy", fontsize=12, weight='bold')
            plt.grid(True, linestyle='--', alpha=0.5)
            plt.legend(frameon=True)

            # --------------------
            # Layout + Save
            # --------------------
            plt.tight_layout(pad=2.0)

            filename = f"{model_name}_{dataset_name}_learning_curve.png"
            filepath = os.path.join(save_dir, filename)

            plt.savefig(filepath, dpi=300, bbox_inches="tight")
            plt.close()

            print(f"Saved: {filepath}")
```

```python
import matplotlib.pyplot as plt
import os

def plot_accuracy_comparison_by_dataset(history_dict, save_dir="plots"):
    os.makedirs(save_dir, exist_ok=True)

    color_map = {
        "DinoV2": "blue",
        "Swin": "red",
        "Vision Transformer": "green"
    }

    datasets = set()
    for model in history_dict:
        datasets.update(history_dict[model].keys())

    for dataset_name in datasets:

        plt.figure(figsize=(8, 5))

        final_accuracy_text = []

        for model_name, dataset_dict in history_dict.items():

            if dataset_name not in dataset_dict:
                continue

            history = dataset_dict[dataset_name]

            epochs = history["epochs"]
            accuracy = history["accuracy"]
            final_accuracy = history.get("final_accuracy", None)

            color = color_map.get(model_name, "black")

            plt.plot(
                epochs,
                accuracy,
                marker='o',
                linewidth=2,
                color=color,
                label=model_name
            )

            if final_accuracy is not None:
                final_accuracy_text.append((model_name, final_accuracy, color))

        # ----------------------------
        # Clean bottom-right colored box
        # ----------------------------
        x_pos = 0.98
        y_start = 0.02
        y_step = 0.06

        for i, (name, acc, color) in enumerate(final_accuracy_text):
            plt.gca().text(
                x_pos,
                y_start + i * y_step,
                f"{name}: {acc:.4f}",
                transform=plt.gca().transAxes,
                fontsize=10,
                horizontalalignment='right',
                verticalalignment='bottom',
                color=color
            )

        plt.title(f"Accuracy Comparison | {dataset_name}", fontsize=13, weight='bold')
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.grid(True, linestyle="--", alpha=0.5)
        plt.legend()

        plt.tight_layout()

        filepath = os.path.join(save_dir, f"{dataset_name}_accuracy_comparison.png")
        plt.savefig(filepath, dpi=300, bbox_inches="tight")
        plt.close()

        print(f"Saved: {filepath}")
```
