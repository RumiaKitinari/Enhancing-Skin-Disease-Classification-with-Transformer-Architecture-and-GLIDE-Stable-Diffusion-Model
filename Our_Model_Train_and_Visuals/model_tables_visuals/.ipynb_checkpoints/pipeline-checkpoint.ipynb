## WARNING: 
`pipeline` is known not to work unless `graphviz` is installed. If you do not have `sudo` access, please delete the `sudo` scripts and run this code on an appropriate local machine. You do not need to import anything; this code simply generates a visualization of the pipeline (already generated for reference). 

```python
%sudo apt-get update
%sudo apt-get install graphviz

from graphviz import Digraph

dot = Digraph("Pipeline", format="png")

# =========================
# GLOBAL STYLE (LIGHT MODE, HIGH CONTRAST)
# =========================
dot.attr(
    rankdir="LR",
    splines="spline",
    nodesep="0.9",
    ranksep="1.2",
    fontsize="11",
    concentrate="true",
    bgcolor="transparent",
    center = "true",
    margin="0.5"
)

dot.attr(
    "node",
    style="filled",
    fontname="Helvetica",
    fontsize="10",
    fontcolor="black",
    color="#000000"
)

dot.attr(
    "edge",
    penwidth="1.3",
    color="#000000"
)

# =========================
# LAYER 1: INPUT
# =========================
input_nodes = ["HAM10000", "ISIC + Atlas"]

with dot.subgraph(name="cluster_input") as c:
    c.attr(
        label="<<B>Input</B>>",
        color="#2F6FB5",
        fontcolor="black",
        style="rounded"
    )
    
    for n in input_nodes:
        c.node(n, n, shape="box3d", fillcolor="#5B9BD5")
    
    c.attr(rank="same")

# =========================
# LAYER 2: AUGMENTATIONS
# =========================
aug_nodes = ["Augmentation", "GLIDE"]

with dot.subgraph(name="cluster_aug") as c:
    c.attr(
        label="<<B>Augmentations</B>>",
        color="#D48806",
        fontcolor="black",
        style="rounded"
    )
    
    for n in aug_nodes:
        c.node(n, n, shape="box3d", fillcolor="#F5A623")
    
    c.attr(rank="same")

# =========================
# LAYER 3: ARCHITECTURE
# =========================
arch_nodes = ["DinoV2", "Swin", "ViT"]

with dot.subgraph(name="cluster_arch") as c:
    c.attr(
        label="<<B>Architecture</B>>",
        color="#2E8B57",
        fontcolor="black",
        style="rounded"
    )
    
    for n in arch_nodes:
        c.node(n, n, shape="box3d", fillcolor="#4CAF50")
    
    c.attr(rank="same")

# =========================
# LAYER 4: RESULTS (as a cluster with single node + 2 invisible nodes)
# =========================
with dot.subgraph(name="cluster_results") as c:
    c.attr(
        label="<<B>Results</B>>",
        color="red",
        fontcolor="black",
        style="rounded"
    )
    
    # Add 2 invisible nodes above and below Results
    c.node("Accuracy", "Accuracy", shape="box3d", fillcolor="red")
    c.node("Performance", "Performance", shape="box3d", fillcolor="red")
    c.node("Results", "Results", shape="box3d", fillcolor="red")
    
    # Stack them vertically
    c.attr(rank="same")
    

# =========================
# LAYER 5: VISUALIZATIONS
# =========================
vis_nodes = ["Tables", "Visualizations", "Graphs"]

with dot.subgraph(name="cluster_vis") as c:
    c.attr(
        label="<<B>Visualizations</B>>",
        color="#9B59B6",
        fontcolor="black",
        style="rounded"
    )
    
    for n in vis_nodes:
        c.node(n, n, shape="box3d", fillcolor="#9B59B6")
    
    c.attr(rank="same")

# =========================
# CONNECTION STYLES
# =========================
input_aug_style = {"color": "#6FA3D9", "penwidth": "1.2"}
aug_arch_style = {"color": "#F0B35A", "penwidth": "1.2"}
arch_res_style = {"color": "#81C784", "penwidth": "1.2"}
res_viz_style = {"color": "#CC2B1C", "penwidth": "1.2"}

# =========================
# CONNECTIONS
# =========================

# Input → Augmentations
for i in input_nodes:
    for a in aug_nodes:
        dot.edge(i, a, **input_aug_style)

# Augmentations → Architecture
for a in aug_nodes:
    for arch in arch_nodes:
        dot.edge(a, arch, **aug_arch_style)

# Architecture → Results (single connection per architecture)
for arch in arch_nodes:
    dot.edge(arch, "Performance", **arch_res_style)
    
# Results → Visualizations
for viz in vis_nodes:
    dot.edge("Performance", viz, **res_viz_style)

# =========================
# RENDER
# =========================
dot.render("pipeline", view=True, cleanup=True)

print("Pipeline diagram generated successfully!")
```
